```python
from tabulate import tabulate

import pandas as pd
movies = pd.read_csv(r"C:\Users\deras\Desktop\AU\data sci\python_palak\Movies_2_Csv.csv")
for i in range (0,len(movies)):
    if movies.Genre[i]=='Drama' or movies.Genre[i]=='Comedy':
        print(tabulate(movies[i], tablefmt = 'psql'))
```

```python
movies.Genre.replace('comedy','Comedy')
```

```python
movies_ascending_Genre = movies.sort_values(by='Genre', ascending = True)
print(movies_ascending_Genre)

movies_ascending_RottenTomatoes = movies.sort_values(by='RottenTomatoes', ascending = True)
print(movies_ascending_RottenTomatoes)
```
